Most of the samples, if ExterQuality is good or excellent, the type is brkface.

```python
def fill_masvnrtype():
    mvt, eq = data["MasVnrType"].values.tolist(),data["ExterQual"].values.tolist()
    for i in range(len(mvt)):
        if type(mvt[i]) == float:
            if eq[i] == "Gd" or eq[i] == "Ex":
                mvt[i] = "BrkFace"
            else:
                mvt[i] = "None"
    data["MasVnrType"] = mvt

fill_masvnrtype()
```

```python
data['BsmtQual'] = data['BsmtQual'].fillna("No Basement")
data['BsmtExposure'] = data['BsmtExposure'].fillna("No Basement")
data['BsmtCond'] = data['BsmtCond'].fillna("No Basement")
data['BsmtFinType1'] = data['BsmtFinType1'].fillna("No Basement")
data['BsmtFinType2'] = data['BsmtFinType2'].fillna("No Basement")
```

```python
def fill_basement():
    bsfsf1, bsft1 = data.BsmtFinSF1.values.tolist(), data.BsmtFinType1.values.tolist()
    bsfsf2, bsft2 = data.BsmtFinSF2.values.tolist(), data.BsmtFinType2.values.tolist()
    for i in range(len(bsft1)):
        if bsft1[i] == "No Basement":
            bsfsf1[i] = 0.0
            bsfsf2[i] = 0.0
            bsft2[i] = "No Basement"
        if bsft2[i] == "No Basement":
            bsfsf1[i] = 0.0
            bsfsf2[i] = 0.0
            bsft1[i] = "No Basement"
    
    data.BsmtFinSF1 = bsfsf1
    data.BsmtFinSF2 = bsfsf2

fill_basement()
```

```python
fill_null("BsmtFinSF1","BsmtFinType1","median")
fill_null("BsmtFinSF2","BsmtFinType2","median")
# fill_null("BsmtFinSF2","BsmtFinType2","mean")
```

Lets create a separate feature which is a sum of the other three basement values(finished1 & 2 and unfinished.)

```python
def fill_basement_total():
    bsfsf1, bsfsf2, bsunf, bstotal = data.BsmtFinSF1.values.tolist(), data.BsmtFinSF2.values.tolist(), data.BsmtUnfSF.values.tolist(), data.TotalBsmtSF.values.tolist()
    
    for i in range(len(bsfsf1)):
        bstotal[i] = bsfsf1[i] + bsfsf2[i] + bsunf[i]
    data.TotalBsmtSF = bstotal

fill_basement_total()
```
